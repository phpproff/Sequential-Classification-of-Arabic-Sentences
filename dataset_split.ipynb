```python
data = '/content/drive/MyDrive/thesis/models/datasetbackup.txt'
```

```python
ioq=0
import re
abstracts = []
abstract = ""
f = open(data, "r", encoding='utf-8-sig',errors="ignore")
for x in f:
    line=re.split(r'\t', x.rstrip('\n'))
    if "#" not in line[0] and line[0] != "":
      abstract = abstract + x   
    if "#" in line[0] and line[0] != "":
      abstracts.append(abstract)
      abstract = ""      
      abstract = abstract + x   
    ''' if ioq > 120:        
      break
    ioq = ioq + 1 '''
abstracts.pop(0)
```

```python
print(abstracts[1])
```

```python
import random
random.shuffle(abstracts)
print(abstracts[1])
```

```python
import math

testing_count = math.ceil( (.1 * len(abstracts))  ) 
validation_count = testing_count

training_count = len(abstracts)  - (testing_count +validation_count)

print("training_count:",training_count,
      ",testing_count:", testing_count,
      ",validation_count:",validation_count)
```

```python
training = abstracts[:training_count]
testing = abstracts[training_count:(training_count+testing_count)]
validation = abstracts[-validation_count:]
```

```python
print("training _count:",len(training),
      ",testing _count:", len(testing),
      ",validation _count:",len(validation))
```
